```python
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
from itertools import chain

#img = cv2.imread("../_fig/_others/miporin.jpg")
img = cv2.imread("../_fig/monet/woman_withparasol.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
height, width, channels = img.shape
print([height, width, channels])
```

```python
croph = 50
cropw = 50
dh = (height-1)/croph
dw = (width-1)/cropw
print([dh, dw])
```

```python
heights=[0]
for i in range(1,croph):
    heights.append(int(np.floor(i * dh)))
    heights.append(int(np.floor(i * dh + 1)))
heights.append(height-1)
```

```python
widths=[0]
for i in range(1,cropw):
    widths.append(int(np.floor(i * dw)))
    widths.append(int(np.floor(i * dw + 1)))
widths.append(width-1)
```

```python
imgF = np.float_(img);
```

```python
imgcrops = []
for ih in range(0,len(heights)-1,2):
    temp = []
    for iw in range(0,len(widths)-1,2):
        temp.append( imgF[heights[ih]:heights[ih+1], widths[iw]:widths[iw+1]] )
    imgcrops.append( temp )
    
imgcrops = np.array(imgcrops)
```

```python
mean_imgs = []
for ih in range(croph):
    temp = []
    for iw in range(cropw):
        imgcrop = imgcrops[ih,iw]
        imgcrop_flatten = np.array(list(chain.from_iterable(imgcrop)))
        temp.append( sum(imgcrop_flatten)/len(imgcrop_flatten) )
    mean_imgs.append(temp)
```

```python
decolors = np.array(mean_imgs)
```

```python
plt.imshow(decolors)
```

```python
decolor_img = np.zeros((2610,2120,3), np.uint8)
#decolor_img = img
for ih in range(0,len(heights)-1,2):
    for iw in range(0,len(widths)-1,2):
        cv2.rectangle(decolor_img, (widths[iw],heights[ih]), (widths[iw+1],heights[ih+1]), decolors[int(ih/2),int(iw/2)], -1, 0)
plt.imshow(decolor_img)
```

```python
def mask_blue(path):
    img = cv2.imread(path)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    blue_min = np.array([100, 170, 200], np.uint8)
    blue_max = np.array([120, 180, 255], np.uint8)

    blue_region = cv2.inRange(hsv, blue_min, blue_max)
    white = np.full(img.shape, 255, dtype=img.dtype)
    background = cv2.bitwise_and(white, white, mask=blue_region)  # detected blue area becomes white

    inv_mask = cv2.bitwise_not(blue_region)  # make mask for not-blue area
    extracted = cv2.bitwise_and(img, img, mask=inv_mask)

    masked = cv2.add(extracted, background)

    return masked
```

```python
img_masked = mask_blue("../_fig/monet/woman_withparasol.jpg")
plt.imshow(img_masked)
```

```python
def mask_color(path):
    img = cv2.imread(path)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    mask_min = np.array([200, 255, 255], np.uint8)
    mask_max = np.array([255, 255, 255], np.uint8)

    mask_region = cv2.inRange(hsv, mask_min, mask_max)
    white = np.full(img.shape, 255, dtype=img.dtype)
    background = cv2.bitwise_and(white, white, mask=mask_region)  # detected mask area becomes white

    inv_mask = cv2.bitwise_not(mask_region)  # make mask for not-mask area
    extracted = cv2.bitwise_and(img, img, mask=inv_mask)

    masked = cv2.add(extracted, background)

    return masked
```

```python
img_masked = mask_color("../_fig/monet/woman_withparasol.jpg")
plt.imshow(img_masked)
```
