```python
import wave
import numpy as np
import struct


def makewave(amp,freq,sec,fs):
    outwave = []
    for i in range(sec*fs):
        w = amp*np.sin(2*np.pi*freq*i/fs)
        outwave.append(w)
    return(outwave)


def wavwrite(inputlist,filename):
    maxamp = 32767.0
    int16wave = [int(x * maxamp) for x in inputlist]
    binwave = struct.pack("h"*len(int16wave),*int16wave)

    nchannnles = 1 # 1=monoral , 2=stereo
    sampwitdth = 2 # 1=8bit,2=16bit,3=,...
    framerate = 44100 # sampling rate ex.44100Hz
    nframes = len(inputlist) # framerate * sec
    
    of = wave.open(filename,"w")
    of.setparams((nchannnles,sampwitdth,framerate,nframes,"NONE","not compressed"))
    of.writeframes(binwave)
    of.close
```

```python
fs = 44100
w1 = makewave(1,800,5,fs)
wavwrite(w1,"sound.wav")
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("../_fig/_others/miporin.jpg",0)
img = cv2.medianBlur(img,5)

ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import numpy as np
import math
import struct
import wave
from itertools import chain
import random

from PIL import Image

import scipy
import scipy.misc #http://docs.scipy.org/doc/scipy/reference/misc.html
import scipy.cluster


# SOUND
def makeWave(amp,freq,sample_freq,sec):
    output_wave = []
    for n in np.arange(sample_freq * sec):
        s = amp * np.sin(2.0 * np.pi * freq * n / sample_freq)
        output_wave.append(s)
    
#        print("HERE")
#        print(s)
#        print(output_wave)

    return(np.array(output_wave))
#    return(output_wave)


def makeSound(color_code):
    tmp = 87.0 * color_code / 255 
    freq_tmp = math.pow(2.0,tmp)
    freq_calc = 27.5 * ((freq_tmp) ** (1.0/12.0))

    sound = makeWave(1,freq_calc,44100,0.5)

    return(sound)


def makeFile_stereo(input_wave,output_filename):

    input_wave = [int(x * 32767.0) for x in input_wave]
    binwave = struct.pack('h' * len(input_wave), *input_wave)
    w = wave.Wave_write(output_filename)
    p = (1, 2, 44100, len(binwave), 'NONE', 'not compressed')

    w.setparams(p)
    w.writeframes(binwave)
    w.close()

    
def makeScore(score,output_filename):
    score = list(chain.from_iterable(score))
    score = np.array(score)    
    makeFile_stereo(score,output_filename)
    

#MAKEMUSIC
def makeMusic(image,output_filename,cropcount):

    score=[]
    image_grid=cropping(image,cropcount)
    
    for i in range(cropcount**2):

        colors_code = decolor(image_grid[i],3,10)
        top_color = colors_code[0]

        top_r = int(colors_code[0][0:2],16)
        top_g = int(colors_code[0][2:4],16)
        top_b = int(colors_code[0][4:6],16)
        
        rsound = makeSound(top_r)
        gsound = makeSound(top_g)
        bsound = makeSound(top_b)

        score.append((rsound + gsound+bsound)/3)
        
    makeScore(score,output_filename)
    
def makeCode(image,cropcount):
#    codes = []
    top_color_decimal = []    
    image_grid=cropping(image,cropcount)

    amp_types = []
    wave_types = []
    t1s = []
    aparams = []
    params = []
    codes_numbers = []
    freqs = []
    
    t1 = 0
    for i in range(cropcount**2):

        colors_code = decolor(image_grid[i],3,10)
        top_color = colors_code[0]

        top_r = int(colors_code[0][0:2],16)
        top_g = int(colors_code[0][2:4],16)
        top_b = int(colors_code[0][4:6],16)

        top_color_decimal = [top_r,top_g,top_b]
        max_rgb = max(top_color_decimal)
        min_rgb = min(top_color_decimal)
        sum_rgb = top_r + top_g + top_b

        codes_number = int(max_rgb/42.5)
        freq = convert_key_code( int(max_rgb/2.89) )
        dt = top_r/255
        t1 = t1 + dt
#        amp_type = int(top_g/85) + 1
#        amp_type = int((top_r + top_g + top_b)/255) + 1
        print([top_r, top_g, top_b, sum_rgb])
        if    sum_rgb < 0 * 3:
            amp_type = 1
        elif  sum_rgb < 0 * 3:
            amp_type = 2
        elif  sum_rgb >= 0 * 3:
            amp_type = 3

#        amp_type = 3

        if   amp_type == 1:
            aparam = [10*top_g]
        elif amp_type == 2:
            aparam = [10*top_g,10*top_g/10]
        elif amp_type >= 3:
            aparam = [0.02, 1.4, 0.08]

#        wave_type = int(top_b/63.75) + 1
#        param = [top_b/255]
        if   min_rgb > 5:
            wave_type = 1
        elif min_rgb <= 5:
            wave_type = 5

        param = [ random.uniform(0,1) ]

        amp_types.append(amp_type)
        wave_types.append(wave_type)
        t1s.append(t1)
 #       t1s = [1,2,3,4]
        aparams.append(aparam)
        params.append(param)
        codes_numbers.append(codes_number)
        freqs.append(freq)

    amp = 1
    sec = 5
    sample_freq = 44100

    codes = [amp_types,wave_types,amp,t1s,aparams,params,codes_numbers,freqs,sample_freq,sec]
    
    print(t1s)
    print(amp_types)
    print(wave_types)
    print(codes_numbers)
    print(freqs)
    #print(codes)

    return(codes)




# amp
def ampFn(amp_type,t,t1,amp,aparam):
    if amp_type == 1:
        ampVal = amp*np.e**(-aparam[0]*(t-t1)**2)

    if amp_type == 2:
        if t <= t1:
            ampVal = amp*np.e**(-aparam[0]*(t-t1)**2)
        elif t > t1:
            ampVal = amp*np.e**(-aparam[1]*(t-t1)**2)

    if amp_type == 3:
        if t <= t1:
            ampVal = 0
        else:
            if t <= aparam[0] + t1:
                ampVal = amp*(t-t1)/aparam[0]
            else:
                ampVal = amp*((aparam[0]+aparam[2])/(t-t1+aparam[2]))**aparam[1]

    return(ampVal)

# Sound
def codewaveFn(wave_type,t,t1,param,codes_number):
    C_freq = 261.63
    D_freq = 293.66
    E_freq = 329.63
    F_freq = 349.23
    G_freq = 392.00
    A_freq = 440.00
    B_freq = 493.88

    if   codes_number == 0:
        code_freq = [C_freq,E_freq,G_freq]
    elif codes_number == 1:
        code_freq = [D_freq,F_freq,A_freq]
    elif codes_number == 2:
        code_freq = [E_freq,G_freq,B_freq]
    elif codes_number == 3:
        code_freq = [F_freq,A_freq,C_freq]
    elif codes_number == 4:
        code_freq = [G_freq,B_freq,D_freq]
    elif codes_number == 5:
        code_freq = [A_freq,C_freq,E_freq]
    elif codes_number == 6:
        code_freq = [B_freq,D_freq,F_freq]

    wave = 0
    for i in np.arange( len(code_freq) ):
        if   wave_type == 1:
            #Sin Wave
            wave = wave + waveFn(1,t,t1,param,code_freq[i])
        elif wave_type == 2:
            #Sawtooth Wave
            wave = wave + waveFn(2,t,t1,param,code_freq[i])
        elif wave_type == 3:
            #Triangle Wave
            wave = wave + waveFn(3,t,t1,param,code_freq[i])
        elif wave_type == 4:
            #Square Wave
            wave = wave + waveFn(4,t,t1,param,code_freq[i])
        elif wave_type == 5:
            #Pulse Wave 1
            wave = wave + waveFn(5,t,t1,param,code_freq[i])

    return(wave)


def waveFn(wave_type,t,t1,param,freq):
    T = 1/freq
    if   wave_type == 1:
        #Sin Wave
        wave = np.sin(2*np.pi * freq*t)
    elif wave_type == 2:
        #Sawtooth Wave
        wave = 2*((t-t1)*freq - np.floor((t-t1)*freq + 0.5))
    elif wave_type == 3:
        #Triangle Wave
        wave = 2*abs(2*((t-t1+T/4)*freq - np.floor((t-t1+T/4)*freq + 0.5)))-1
    elif wave_type == 4:
        #Square Wave
        wave = np.sign(np.sin(2*np.pi * freq*t))
    elif wave_type == 5:
        #Pulse Wave 1
        wave = np.sign(- (t-t1)*freq + np.floor((t-t1)*freq) + param[0])

    return(wave)


def makeSyntheticWave(amp_types,wave_types,amp,t1s,aparams,params,codes_numbers,freqs,sample_freq,sec):
    output_wave = []
    output_wave_abs = []

    for it in np.arange(sample_freq * sec):
        t = it / sample_freq  # time
        w = 0
        for i in np.arange(len(t1s)):
            w = w + ampFn(amp_types[i],t,t1s[i],amp,aparams[i]) * codewaveFn(wave_types[i],t,t1s[i],params[i],codes_numbers[i])
            w = w + ampFn(amp_types[i],t,t1s[i],amp,aparams[i]) * waveFn(wave_types[i],t,t1s[i],params[i],freqs[i])

        output_wave.append(w)
        absw = abs(w)
        output_wave_abs.append(absw)
    print( max(output_wave_abs) )
    wmax = max(output_wave_abs)

    for it in np.arange(sample_freq * sec):
        output_wave[it] = 0.99*output_wave[it]/wmax
    print( max(output_wave) )

#    return(np.array(output_wave))
    return(output_wave)


def convert_key_code(key_code):
    tmp = key_code 
    freq_tmp = math.pow(2.0,tmp)
    freq_calc = 27.5 * ((freq_tmp) ** (1.0/12.0))
    return(freq_calc)
```
